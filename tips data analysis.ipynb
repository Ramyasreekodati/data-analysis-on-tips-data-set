```python
continues = ["total_bill", "tip"]  # List of continuous variables to analyze.
quntity = [0, 0.25, 0.5, 0.75, 1]  # List of quantiles to calculate.

# Loop through each continuous variable
for i in continues:
    # Loop through each quantile in the list
    for j in quntity:
        d = df[i].quantile(j)  # Calculate the j-th quantile for the current variable.
        print(f"{i} the quantile of ({j}) is    : {d}")  # Print the quantile value.

    q1 = df[i].quantile(0.25)  # Calculate the first quartile (25th percentile).
    q3 = df[i].quantile(0.75)  # Calculate the third quartile (75th percentile).
    iqr = q3 - q1  # Calculate the interquartile range (IQR).
    print("\n")
    print(f"{i} the interquartile range (IQR) is :  {iqr}")  # Print the IQR.

    ll = q1 - (1.5 * iqr)  # Calculate the lower limit for outlier detection.
    ul = q3 + (1.5 * iqr)  # Calculate the upper limit for outlier detection.
    print(f"lowerlimit is {ll} and upperlimit is {ul}")  # Print the lower and upper limits.

    outliers = df[(df[i] < ll) | (df[i] > ul)]  # Identify outliers based on the limits.
    print("Outliers:")  # Print a heading for the outliers.
    print(outliers)  # Print the outliers found in the DataFrame.

    filtered_df = df[(df[i] >= q1) & (df[i] <= q3)]  # Filter the DataFrame to exclude outliers.
    shape_after_iqr = filtered_df.shape  # Get the shape (number of rows and columns) of the filtered DataFrame.
    print(f"Shape after applying IQR: {shape_after_iqr}")  # Print the shape of the filtered DataFrame.
```
